# Introducción a Matplotlib
Matplotlib es una biblioteca de visualización de datos para Python que permite crear gráficos estáticos, animados e interactivos de alta calidad en pocas líneas de código. Soporta una amplia variedad de gráficos y formatos de visualización, incluyendo gráficos de líneas, histogramas, diagramas de dispersión, gráficos de barras, gráficos de error, diagramas de caja, mapas de calor, y mucho más.  
En [este link](https://matplotlib.org/stable/tutorials/pyplot.html) encontraran una pequeña introducción dentro de la propia documentación oficial de matplotlib.
___
## Gráficos de líneas
Los gráficos de línea utilizan una línea continua para conectar puntos de datos que representan valores secuenciales de una o más variables, mostrando tendencias o cambios a lo largo del tiempo.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 50, 1000)
y = np.sin(2*x) * np.exp(-0.1*x)
```

```python
plt.plot(x, y, color= 'red', label= 'Función Senoidal Atenuada')
plt.plot(x, np.exp(-0.1*x), label= 'Envolvente positiva', linestyle= ':')
plt.plot(x, -np.exp(-0.1*x), label= 'Envolvente negativa', ls= '-.')

plt.title('Gráfico de una función atenuada')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud [m]')
plt.legend()
plt.grid()

plt.show()
```

## Gráficos de dispersión
Los gráficos de dispersión son visualizaciones que utilizan puntos en un sistema de coordenadas para representar los valores de dos variables, permitiendo observar la relación o correlación entre ellas.  
En [este link](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html) podés encontrar información relevante para profundizar en el tema.

```python
x = np.arange(50)
y = x**2 + np.random.randint(50, 250, 50)
```

```python
plt.scatter(x, y)
plt.title('Gráfico de disperción', size= 16)
plt.xlabel('Tiempo')
plt.ylabel('Crecimiento')

plt.show()
```

## Histogramas
Los histogramas son gráficos que muestran la distribución de frecuencias de un conjunto de datos, agrupando los datos en intervalos o "bins" y representando la cantidad de datos en cada bin con barras.  
En [este link](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html) podés encontrar información relevante sobre histogramas.

```python
x = np.random.randn(1000)
z = np.random.randn(1000) + 3
```

```python
plt.hist(x, color= 'green', alpha= 0.5, bins=5)
plt.hist(z, color= 'blue', alpha= 0.5)
plt.title('Gráfico de Histograma', size= 16)
plt.xlabel('Datos')
plt.ylabel('Frecuencia')

plt.show()
```

## Heatmap
Un mapa de calor es una representación gráfica de datos donde los valores individuales contenidos en una matriz se representan como colores.  
En [este link](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imshow.html) podés encontrar información relevante sobre heatmaps.

```python
datos = np.random.randn(20,20)
datos.shape
```

```python
plt.imshow(datos, cmap= 'hot')

plt.show()
```
